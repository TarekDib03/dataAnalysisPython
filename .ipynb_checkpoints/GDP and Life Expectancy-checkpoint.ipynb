Richer countries can afford to invest more on healthcare, on job and road safety, and other measures that could reduce mortality. On the other hand, richer countries may have less healthy lifestyles. Is there any relation between the wealth of a country and the life expectancy of its inhabitants?

The following analysis checks whether there is any correlation between the total gross domestic product (GDP) of a country in 2013 and the life expectancy of people born in that country in 2013.

## Getting the data

Two datasets of the World Bank are considered. One dataset, available at <http://data.worldbank.org/indicator/NY.GDP.MKTP.CD>, lists the GDP of the world's countries in current US dollars, for various years. The use of a common currency allows us to compare GDP values across countries. The other dataset, available at <http://data.worldbank.org/indicator/SP.DYN.LE00.IN>, lists the life expectancy of the world's countries. The datasets were downloaded using pandas.io.wb package on June 26, 2016.

```python
import warnings
warnings.simplefilter('ignore', FutureWarning)

from pandas import *
from pandas.io.wb import download

YEAR = [2010,2015]
GDP_INDICATOR = 'NY.GDP.MKTP.CD'
gdpWB = download(indicator=GDP_INDICATOR, country='all', start=YEAR[0], end=YEAR[1])

LIFE_INDICATOR = 'SP.DYN.LE00.IN'
lifeWB = download(indicator=LIFE_INDICATOR, country='all', start=YEAR[0], end=YEAR[1])
lifeWB.head()
```

In the above dataframe, country and year are indexes. Reseting the index puts the dataframe in its normal form.

```python
gdpReset = gdpWB.reset_index()
lifeReset = lifeWB.reset_index()
lifeReset[204:210]
```

- GDP and life expectancy values are missing for some countries.

1. removing the first 34 rows;
- removing rows with unavailable values.

```python
# Drop null values from the dataframes.
gdpCountriesClean = gdpReset[204:].dropna()
lifeCountriesClean = lifeReset[204:].dropna()
lifeCountriesClean.dtypes
```

## GDP and Life Expectancy Data, 2013

I will focus on the Year 2014 for the rest of this report.

```python
YEAR = 2013
# GDP
GDP_INDICATOR = 'NY.GDP.MKTP.CD'
gdpWB = download(indicator=GDP_INDICATOR, country='all', start=YEAR, end=YEAR)

# Life Expectancy
LIFE_INDICATOR = 'SP.DYN.LE00.IN'
lifeWB = download(indicator=LIFE_INDICATOR, country='all', start=YEAR, end=YEAR)

# Population
POPULATION_INDICATOR = "SP.POP.TOTL"
populationWB = download(indicator = POPULATION_INDICATOR, country = "all", start = YEAR, end = YEAR)

# Reset index
gdpReset = gdpWB.reset_index()
lifeReset = lifeWB.reset_index()
populationReset = populationWB.reset_index()

gdpCountries = gdpReset[34:]
lifeCountries = lifeReset[34:]
populationCountries = populationReset[34:]
gdpCountries.head()
```

```python
# Round GDP to millions
def roundToMillions (value):
    return round(value / 1000000)

# GDP per capita
def gdpPerCapita(gdp, pop):
    return gdp / pop

GDP = 'GDP ($m)'
gdpCountries[GDP] = gdpCountries[GDP_INDICATOR].apply(roundToMillions)
gdpCountries.head()
```

The unnecessary columns can be dropped.

```python
COUNTRY = 'country'
headings = [COUNTRY, GDP]
gdpClean = gdpCountries[headings]
gdpClean.head()
```

The World Bank reports the life expectancy with several decimal places. After rounding, the original column is discarded.

```python
LIFE = 'Life expectancy (years)'
lifeCountries[LIFE] = lifeCountries[LIFE_INDICATOR].apply(round)
headings = [COUNTRY, LIFE]
lifeClean = lifeCountries[headings]
lifeClean.head()
```

```python
POPULATION = 'Population'
headings = [COUNTRY, POPULATION]
populationCountries[POPULATION] = populationCountries[POPULATION_INDICATOR]
headings = [COUNTRY, POPULATION]
popClean = populationCountries[headings]
popClean.head()
```

```python
# Any null values?
popClean.isnull().values.any()
```

## Combining the data

The tables are combined through an inner join on the common 'country' column. 

```python
gdpVsLife = merge(gdpClean, lifeClean, on=COUNTRY, how='inner').dropna()
gdpVsLifeVsPop = merge(gdpVsLife, popClean, on = COUNTRY, how='inner')

GDP_PER_CAPITA = 'GDP ($)/Capita'
gdpVsLifeVsPop[GDP_PER_CAPITA] = (gdpVsLifeVsPop[GDP]*1000000 / gdpVsLifeVsPop[POPULATION]).apply(round)
gdpVsLifeVsPopSorted = gdpVsLifeVsPop.sort(GDP_PER_CAPITA, ascending = True)
gdpVsLifeVsPopSorted.head(10)
```

```python
gdpVsLifeVsPopSorted.tail(10).sort(GDP_PER_CAPITA, ascending = False)
```

```python
# The 10 countries with the highest GDP
gdpTop10 = gdpVsLife.sort('GDP ($m)', ascending=False).head(10)
gdpTop10
```

```python
LifeTop10 = gdpVsLife.sort('Life expectancy (years)', ascending=False).head(10)
LifeTop10
```

```python
# Join the above the data sets
frames = [gdpTop10, LifeTop10]
gdpVsLifeTop10 = merge(gdpTop10, LifeTop10, on=COUNTRY, how='inner')
#gdpVsLifeTop10

# Countries in the top 10 of GDP and life expectancy
gdpVsLifeTop = gdpVsLifeTop10[['country', 'GDP ($m)_x', 'Life expectancy (years)_x']]
gdpVsLifeTop.rename(columns={'GDP ($m)_x':'GDP($m)', 'Life expectancy (years)_x':'Life expectancy (years)'}, inplace=True)
gdpVsLifeTop
```

```python
frames = [gdpTop10, LifeTop10]
gdpVsLifeTopCountries = concat(frames)
gdpVsLifeTopCountries.sort('Life expectancy (years)', ascending = True)
```

## Calculating the correlation

To measure if the life expectancy and the GDP grow together, the Spearman rank correlation coefficient is used. A perfect correlation doesn't imply any cause-effect relation between the two indicators. A p-value below 0.05 means the correlation is statistically significant.

```python
from scipy.stats import spearmanr

gdpPerCapitaColumn = gdpVsLifeVsPop[GDP_PER_CAPITA]
lifeColumn = gdpVsLifeVsPop[LIFE]
(correlation, pValue) = spearmanr(gdpColumn, lifeColumn)
print('The correlation is', correlation)
if pValue < 0.05:
    print('It is statistically significant.')
else:
    print('It is not statistically significant.')
```

The value shows a direct correlation, i.e. richer countries tend to have longer life expectancy, but it is not very strong.

## Showing the data

Measures of correlation can be misleading, so it is best to see the overall picture with a scatterplot. The GDP axis uses a logarithmic scale to better display the vast range of GDP values, from a few million to several billion (million of million) pounds.

```python
%matplotlib inline
gdpVsLifeVsPop.plot(x=GDP_PER_CAPITA, y=LIFE, kind='scatter', grid=True, logx = True, figsize=(15, 5))
```

From the Figure above, there does not seem to be a clear correlation between GDP/Capita and life expectancy. There are few countries with relatively high GDP/Capita but low life expectancy. On the other hand, there are some countries with relatively low GDP/Capita but high life expectancy.

## GDP Per Capita Based on Purchasing Power Parity
GDP per capita based on purchasing power parity (PPP). PPP GDP is gross domestic product converted to international dollars using 
purchasing power parity rates. An international dollar has the same purchasing power over GDP as the U.S. dollar has in the 
United States. GDP at purchaser's prices is the sum of gross value added by all resident producers in the economy plus any 
product taxes and minus any subsidies not included in the value of the products. It is calculated without making deductions 
for depreciation of fabricated assets or for depletion and degradation of natural resources.

```python
GDP_PER_CAPITA_INDICATOR = 'NY.GDP.PCAP.PP.CD'
gdpPerCapitaWB = download(indicator=GDP_PER_CAPITA_INDICATOR, country='all', start=YEAR, end=YEAR)
gdpPerCapitaReset = gdpPerCapitaWB.reset_index()
gdpPerCapitaCountries = gdpPerCapitaReset[34:]
GDP_PER_CAPITA = 'GDP ($) / Capita'
gdpPerCapitaCountries[GDP_PER_CAPITA] = gdpPerCapitaCountries[GDP_PER_CAPITA_INDICATOR].apply(round)
headings = [COUNTRY, GDP_PER_CAPITA]
gdpPerCapitaClean = gdpPerCapitaCountries[headings].dropna()
gdpPerCapitaClean.head()
```

```python
# Merge data sets
gdpVsLifeVsPerCapita = merge(gdpVsLife, gdpPerCapitaClean, on = COUNTRY, how = 'inner')
gdpVsLifeVsPerCapita.head()
```

```python
# Estimate Correlation between GDP/Capita and life expectance
gdpPerCapitaColumn = gdpVsLifeVsPerCapita[GDP_PER_CAPITA]
lifeColumn = gdpVsLifeVsPerCapita[LIFE]
(correlation, pValue) = spearmanr(gdpPerCapitaColumn, lifeColumn)
print('The correlation is', correlation)
if pValue < 0.05:
    print('It is statistically significant.')
else:
    print('It is not statistically significant.')
```

There is  astrong correlation between GDP per Capita and life expectancy.

```python
gdpVsLifeVsPerCapita.plot(x=GDP_PER_CAPITA, y=LIFE, kind='scatter', grid=True, logx = True, figsize=(15, 5))
```

```python
# The 10 countries with the highest GDP per Capita
gdpVsLifeVsPerCapita.sort(GDP_PER_CAPITA, ascending = False).head(10)
```

## Conclusions

To sum up, there is no strong correlation between a country's wealth and the life expectancy of its inhabitants: there is often a wide variation of life expectancy for countries with similar GDP, countries with the lowest life expectancy are not the poorest countries, and countries with the highest LIFE expectancy are not the richest countries. Nevertheless there is some relationship, because the vast majority of countries with a life expectancy below 70 years is on the left half of the scatterplot.

Using the [NY.GDP.PCAP.PP.CD](http://data.worldbank.org/indicator/NY.GDP.PCAP.PP.CD) indicator, GDP per capita in current 'international dollars', made for a better like-for-like comparison between countries, because it took population and purchasing power into account.
